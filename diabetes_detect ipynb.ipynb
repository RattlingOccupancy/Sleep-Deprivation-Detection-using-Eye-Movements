```python
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
diabetes_data= pd.read_csv("/content/diabetes.csv")
```

```python
x= diabetes_data.drop(columns = "Outcome", axis= 1)
y= diabetes_data["Outcome"]
```

```python
x
```

```python
y
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardised_data = scaler.transform(x)
```

```python
standardised_data
```

```python
x= standardised_data
y= diabetes_data["Outcome"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, stratify=y, random_state =2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
classifier = svm.SVC(kernel="linear")
```

```python
classifier.fit(x_train, y_train)
```

```python
x_prediction = classifier.predict(x_train)
x_predict_accuracy = accuracy_score(x_prediction, y_train)
```

```python
print("the prediction accuracy is:",x_predict_accuracy*100,"%")
```

```python
x_test_predict = classifier.predict(x_test)
x_test_predict_accuracy = accuracy_score(x_test_predict, y_test)
```

```python
print("the test prediction accuracy is:",x_test_predict_accuracy*100,"%")
```

```python
input_data= (7,106,92,18,0,22.7,0.235,48) #taken from csv dataset

input_data_as_np_array = np.asarray(input_data)

input_data_reshape = input_data_as_np_array.reshape(1,-1)

input_data_standard_scaler = scaler.transform(input_data_reshape)

input_data_predict = classifier.predict(input_data_standard_scaler)

print(input_data_predict)

if(input_data_predict[0] ==0 ):
  print("the person is non diabetic")
else:
  print("the person is diabetic")
```
